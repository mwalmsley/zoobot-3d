```python
df = pd.read_csv('/Users/user/repos/zoobot-3d/data/gz3d_and_gz_desi_master_catalog.csv')
galaxy = df.iloc[32]
galaxy
```

```python
# index 9 is spiral arm responses
# index 10 is bar paths
raw_responses = fits.open(galaxy['local_gz3d_fits_loc'])[9].data
raw_responses
```

```python
dimensions=(525, 525)
coords = [[x, y] for y in range(dimensions[1]) for x in range(dimensions[0])]

fig, (ax0, ax1, ax2) = plt.subplots(ncols=3, figsize=(10, 5))
mask = np.zeros(dimensions)
for user_components in spiral_marks_by_user:
    mask += segmap_utils.draw_components(user_components, remove_self_intersecting=False)
ax0.imshow(mask)
ax0.axis('off')

mask = np.zeros(dimensions)
for user_components in spiral_marks_by_user:
    mask += segmap_utils.draw_components(user_components, remove_self_intersecting=True)
ax1.imshow(mask)
ax1.axis('off')

ax2.imshow(fits.open(galaxy['local_gz3d_fits_loc'])[0].data[::-1])
ax2.axis('off')
```
